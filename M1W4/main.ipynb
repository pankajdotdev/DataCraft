```python

df = pd.read_csv("../Datasets/Titanic-Dataset-Cleaned.csv")
df 
```

```python
sns.countplot(x='Survived', data=df)
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
```

```python
cols = ['Sex', 'Pclass', 'Embarked']

for col in cols:
    sns.countplot(x=col, hue='Survived', data=df)
    plt.title(col)
    plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize=(15,4))

sns.countplot(x='Sex', data=df, ax=axes[0])
sns.countplot(x='Pclass', data=df, ax=axes[1])
sns.countplot(x='Embarked', data=df, ax=axes[2])

plt.show()
```

```python
sns.countplot(x='Survived', data=df, palette='Set1')
```

## Kitne passengers survive hue aur kitne nahi hue?

## Passengers ki age distribution kya hai?

```python
sns.histplot(x='Age', data=df, bins=30)
```

## Higher fare wale passengers zyada survive hue kya?

```python
sns.boxplot(x='Survived', y='Fare', data=df)
```

```python
sns.scatterplot(x='Age', y='Fare', data=df, hue='Survived')
```
